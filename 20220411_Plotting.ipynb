```python
import numpy as np
import pandas as pd
from numpy.random import randint
```

```python
df1 = pd.DataFrame(randint(10, 20, (3, 4)), index = list('abc'), columns=[list('ABCD'), list('XXYY')])
print(df1)
```

```python
g1 = df1.groupby(axis=1, level=1)

# Same thing
print(g1.sum()) # Most simple
print(g1.agg(np.sum)) # Modify with known functions or my DIY functions
print('-'*15)

def myfunc(df_sub):
  r, c = df_sub.shape
  ret = pd.Series(np.zeros(r), index = df_sub.index)
  for i in df_sub:
    ret += df_sub[i]
  return ret

print(g1.agg(myfunc)) # Costumizing
```

# Plotting (Series)

```python
s1= pd.Series(np.random.randn(1000))
print(s1)

x1 = s1.cumsum().plot() # matplotlib.axes._subplots
print(type(x1))
```

# Plotting (DataFrame)

```python
df1 = pd.DataFrame(np.random.randn(1000, 4), columns = list('ABCD'))
print(df1)
df1.cumsum().plot()
# Make label and choose color for each Series automatically
```

# Choose each axis values
- plot(x = [], y = [])

```python
df1.cumsum().plot(y=['A', 'B'])
df1.cumsum().plot(x='A', y='B')
```

# Select graph type
- plot(kind =    )
- List of kinds:
  - **‘line’** : line plot (default)
  - **‘bar’** : vertical bar plot
  - **‘barh’** : horizontal bar plot
  - **‘hist’** : histogram
  - **‘box’** : boxplot
  - **‘kde’** : Kernel Density Estimation plot
  - **‘density’** : same as ‘kde’
  - **‘area’** : area plot
  - **‘pie’** : pie plot
  - **‘scatter’** : scatter plot (DataFrame only)
  - **‘hexbin’** : hexbin plot (DataFrame only)

```python
df1 = pd.DataFrame(randint(0, 100, (3, 4)), index = list('abc'), columns = list('ABCD'))
print(df1)
```

```python
df1.plot(kind = 'bar', title='Bar plot')
df1.plot(kind = 'bar', title='Bar plot', stacked=True)
```

```python
df1.plot(kind = 'barh', title='Barh plot')
df1.plot(kind = 'barh', title='Barh plot', stacked = True)
```

```python
df1['A'].plot(kind = 'hist', title='Histogram', width = 0.4)
```

```python
df1.plot(kind = 'density', title='Density')
```

```python
df1.plot(kind = 'area', title='Area') # Similar with stacked = True
```

```python
df1.plot(x = 'A', y = 'B', kind = 'scatter', title='Scatter plot')
```

```python
df1.plot(x = 'A', y = 'B', kind = 'hexbin', title='Hexbin plot')
```

```python
df1.loc['a'].plot(kind = 'pie', title='Pie chart')
```

```python
df1.plot(kind = 'box', title='Box plot')
```

```python
df1.plot(kind = 'kde', title='Kernel Density Estimation plot')
```
