# 2D array, indexing, basic operations with **Numpy**

```python
import numpy as np
import matplotlib as plt

```

```python
def print_info(obj):
  print('type: ', type(obj))
  print('print: ')
  print(obj)

  print('ndim: ', obj.ndim)
  print('shape: ', obj.shape)
  print('size: ', obj.size)
  print('dtype: ', obj.dtype) # dtype: Data type
  print('itemsize: ', obj.itemsize) # itemsize: Number of bytes that one of element occupied
  print('data: {}'.format(obj.data))  # data: Address of array
  print('-'*20)
```

```python
a1 = np.array([[1,2,3], [4,5,6]])
print_info(a1)
```

# Creation of 2D array using Numpy

# reshape

```python
a2 = np.arange(10, 20)
print_info(a2)

print(a2.reshape(2, 5))
a2 = a2.reshape(2, 5)
print(a2)

print(a2.reshape(5,-1)) # -1 means auto calculation
```

```python
# From 25 to 36, # of numbers = 12, type = ndarray
# Shape (2, 6)  / (3, 4)  / (4, 3)

a3 = np.linspace(25, 36, 12)
print_info(a3.reshape(2,-1)) # shape = (2, 6)
print_info(a3.reshape(3,-1)) # shape = (3, 4)
print_info(a3.reshape(4,-1)) # shape = (4, 3)

```

# fromfunction

```python
def fn(a, b): # input: index value
  return a*5 + b

a1 = np.fromfunction(fn, (5, 3), dtype = np.float64)
print(a1, '\n')

a2 = np.fromfunction(fn, (2, 5), dtype = np.float64)
print(a2)
 
```

# np.arange

```python
x = np.arange(-10, 10)
y = np.arange(5, 25)
# x and y must have same number of elements

a3 = fn(x, y).reshape(5, -1)
print(a3)
```

```python
def fn2(a): # input: index value
  return a*5 + a

x = np.arange(-10, 10)
y = np.zeros_like(x)

#for i, v in enumerate(x):
  #y[i] = fn2(v)

y = fn2(x)   # with using function
y2 = x*5 + x # without using function

print(x)
print(y)
print(y2)
```

# eye, identity, full
- eye(N, M, S): Create Matrix that shape(N x M) 

```python
a1 = np.eye(5)
print(a1)

a2 = np.eye(3,5)
print(a2)

a3 = np.eye(3, 5, 2) 
print(a3)
```

- np.identity(N): Create Matrix (N x N)

```python
a1 = np.identity(5)
print(a1)
```

- np.full(**shape**, **value**): Create Matrix that have shape and all values = value

# Stacking
- np.vstack((array, array))  
  - Merge two matrix in **vertical way**

- np.hstack((array, array))  
  - Merge two array in **horizontal way**


```python
a1 = np.arange(10 ,15)
a2 = np.arange(20, 25)
print(a1)
print(a2)

a3 = np.vstack((a1, a2))
print(a3)
a3 = np.hstack((a1, a2))
print(a3)
```

```python
a1 = np.arange(10, 19).reshape(3,3)
a2 = np.arange(20, 29).reshape(3,3)

a3 = np.vstack((a1, a2))
a4 = np.hstack((a1, a2))
print(a1)
print('-'*20)
print(a2)
print('-'*20)
print(a3)
print('-'*20)
print(a4)
```

# Slicing
- np.vsplit(array, number) $~~~$ [Opposite of vstack]
  - Split array to number in **vertical way**
  - When use **vstack**(two arrays) then, reform to original array  
  

- np.hstack(array, number)  $~~~$ [Opposite of hstack]
  - Split array to number in **horizontal way**
  - When use **hstack**(two arrays) then, reform to original array

- Output: **List** (Not ndarray)

  image.png

  image.png

Reference: https://www.w3resource.com/numpy/manipulation/vsplit.php

```python
a1 = np.arange(10, 50).reshape(4, 10)
print(a1)

a2 = np.vsplit(a1, 2)
a3 = np.hsplit(a1, 2)
print_info(np.array(a2))
print_info(np.array(a3))

print(np.hsplit(a1, (3, 7, 8))) # tuple represent splicing index

print(np.vsplit(a1, (1, 3))) # tuple represent splicing index
```

# Indexing

```python
a1 = np.arange(100, 200, 2).reshape(5, -1)
print(a1)
print(a1[1, 1])
print(a1[:2, :3])
print(a1[1:3, 3:5])
print('-'*20)
print(a1[::3, ::3])
```

Integer

```python
print(a1[1:4, 6:9])
print('-'*20)
print(a1[:,5:7])
```

Integer list

```python
print(a1)
print(a1[[1,3,4], [3,5,8]]) # index [1, 3], [3, 5], [4, 8]
print(a1[:, [3, 5, 8]])
```

Boolen list

```python
print(a1)
#print(a1 % 3 == 0)
print(a1[a1 % 3 == 0])

#print(a1 % 5 == 0)
print(a1[a1 % 5 == 0])
print(a1[(a1 % 3 == 0) & (a1 % 5 == 0)])
print(a1[(a1 % 3 == 0) | (a1 % 5 == 0)])
```

**np.diag**: get diagonalized indexed elements

```python
print(a1)
print(np.diag(a1))
print(np.diag(a1, 2))
```

**np.tril**: get elements lower than diagonal

```python
print(np.tril(a1))  # lower than diagonal
print('-'*50)
print(np.tril(a1, 5))
```

**np.triu**: get elements upper than diagonal

```python
print(np.triu(a1)) # upper than diagonal
print('-'*50)
print(np.triu(a1, 5))
```

```python
print(np.tri(3, 5))
print(np.tri(3, 5, 2))
```

# Basic operations

```python
# a1 From 10 to 19, 2D array, (2, -1)
a1 = np.arange(10 ,20).reshape(2, -1)

# a2 from -10 to -19, 2D array (2, -1)
a2 = np.arange(-10, -20, -1).reshape(2, -1)

print(a1)
print(a2)
print(a1 + a2)
print(a1 - a2)
print(a1 * a2)
print(a1 / a2)
print(a1 // a2)
print(a1 % a2)
```

# Axis

np.newaxis: Create new axis in that position

```python
a1 = np.array([1,2,3,4,5])
print(a1)

a2 = a1[np.newaxis, :]
print_info(a2)

a3 = a1[:, np.newaxis]
print_info(a3)
```

```python
a1 = np.arange(20).reshape(2,-1)
print(a1)
print(a1.sum())

print(a1.shape)
print('axis 0: ', a1.sum(axis=0)) # Sum in shape[0]
print('axis 1: ', a1.sum(axis=1)) # Sum in shape[1]
```
